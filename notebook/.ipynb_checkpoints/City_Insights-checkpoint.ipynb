```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/City_Subset_with_TimeSeries.csv")

df.order_date = pd.to_datetime(df.order_date,format='%Y-%m-%d') 

number_of_weeks = max(df.order_week)
number_of_months = max(df.order_month)
date_range = int(str(max(df.order_date) - min(df.order_date)).split(" ")[0])
df["week_of_month"] = df.order_day.apply(lambda x: np.ceil(x/7))
df.head()
```

```python
df["week_of_month"] = df.order_day.apply(lambda x: np.ceil(x/7))
```

```python
city_insights = {}
city_insights["num_users"] = len(df.customer_id.unique())
consolidated = df.groupby("order_id")
city_insights["mean_order_cost"] = consolidated.total_cost.sum().mean()
city_insights["avg_value_per_day"] = (df.groupby("order_date").total_cost.sum().sum())/date_range
city_insights["avg_order_per_day"] = (df.groupby("order_date").order_id.count().sum())/date_range
city_insights["number_of_warehouses"] = df.groupby(["city_id","store_id"]).count().shape[0]
city_insights["day_of_week_graph"] = [i/number_of_weeks for i in df.groupby("order_day_of_week")["total_cost"].sum()]
city_insights["week_of_month_graph"] = [i/number_of_months for i in df.groupby("week_of_month")["total_cost"].sum()]

city_insights = pd.DataFrame([city_insights])
city_insights
```

```python
import geopy.distance

def getDistance(lat1, long1, lat2=warehouse['lat'], long2=warehouse['long']):
    return geopy.distance.distance((lat1, long1), (lat2, long2)).km
```

```python
distances = []
for center in Area_centers:
    distances.append(getDistance(center["lat"],center["long"]))
city_insights["max_distance_from_warehouse"] = max(distances)
city_insights["warehouse_lat"] = warehouse["lat"]
city_insights["warehouse_long"] = warehouse["long"]
```

### Look at Subscription

```python


y = df.groupby(["order_date"]).total_cost.sum()
df.drop(['customer_id', 'manufacturer_id', 'society_id', 'city_id', 'route_id',
       'store_id', 'order_id', 'category_id', 'subcategory_id','order_placed_day','order_placed_month','order_placed_day_of_week','order_placed_hour',
       'product_id', 'product_quantity', 'selling_price_per_unit','subscription', 'product_addedtobasket_on'],inplace=True,axis=1)
x = df.groupby(["order_date"]).mean()
```

```python
x = x.reset_index()
x["time_value"] = x.index
```

```python
x.drop(["order_date","total_cost"],axis=1,inplace=True)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, shuffle=False )
```

```python
xgb_reg = xgb.XGBRegressor(n_estimators=200, n_jobs=-1,max_depth=3,verbosity =0)
m=xgb_reg.fit(X_train, y_train)
pred = m.predict(X_test)
```

```python
math.sqrt(mean_squared_error(y_test, pred))
```

```python
m=xgb_reg.fit(x,y)
```

### Alpha_Centers

```python
pods = pd.read_csv("../output/pod_data.csv")
pod1 = pods["pod1"].idxmin()
pod2 = pods["pod2"].idxmin()
pod3 = pods["pod3"].idxmin()

city_insights["pod1_lat"] = pods.lat[pod1]
city_insights["pod1_long"] = pods.long[pod1]
city_insights["pod2_lat"] = pods.lat[pod2]
city_insights["pod2_long"] = pods.long[pod2]
city_insights["pod3_lat"] = pods.lat[pod3]
city_insights["pod3_long"] = pods.long[pod3]
city_insights.to_csv("../output/city_descriptive_insights.csv",index=None)
```

```python
pods["distance"] = pods.apply(lambda x: min(x.pod1,x.pod2,x.pod3),axis=1)

city_insights["predicted_distance"] = pods.distance.max()
```
