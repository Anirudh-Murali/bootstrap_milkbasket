```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/City_Subset_with_TimeSeries.csv")
society_lat_long = pd.read_csv("../data/societies_latlong.csv")
# society_lat_long.head()
society_lat_long = society_lat_long.set_index("society_id")
# df["lat"] = df.society_id.apply(lambda x: society_lat_long["lat"][x])
# df["long"] = df.society_id.apply(lambda x: society_lat_long["long"][x])
df["area"] = df.society_id.apply(lambda x: society_lat_long["area"][x])

df.order_date = pd.to_datetime(df.order_date,format='%Y-%m-%d') 

number_of_weeks = max(df.order_week)
number_of_months = max(df.order_month)
date_range = int(str(max(df.order_date) - min(df.order_date)).split(" ")[0])
df["week_of_month"] = df.order_day.apply(lambda x: np.ceil(x/7))
df.head()
```

```python
prod = pd.DataFrame(df.product_id.unique(),columns=["product_id"])
prod.to_csv("product_ids.csv",index=None)
```

```python
customer_q = pd.read_csv("../output/customer_intermediate.csv")
# area_q.set_index("area")
customer_q
```

```python

# areas = df.area.unique()
customer_data = []
for customer in customer_q.customer_id:
    customer_insights = {}
    customer_df = df.loc[df.customer_id == customer]
    customer_insights["customer_id"] = customer
    customer_insights["num_products"] = len(customer_df.product_id.unique())
    if customer_df.subscription.any()==1:
        customer_insights["subscription"]  = 1
    else:
        customer_insights["subscription"]  = 0
#     customer_insights["num_societies"] = len(area_df.society_id.unique())
    customer_insights["quintile"] = customer_q.loc[customer_q.customer_id == customer].quintile.reset_index()["quintile"][0]
    customer_insights["top_5_products"] = [i for i in customer_df.groupby(["product_id"]).sum().sort_values("product_quantity",ascending=False).index[:5].values]
#     customer_insights["top_5_brands"] = [i for i in customer_df.groupby(["manufacturer_id"]).sum().sort_values("product_quantity",ascending=False).index[:5].values]
    consolidated = customer_df.groupby("order_id")
    customer_insights["mean_order_cost"] = consolidated.total_cost.sum().mean()
    customer_insights["favourite_brand"] = customer_df.manufacturer_id.mode().values[0]
    customer_insights["product_category"] = customer_df.category_id.mode().values[0]
    customer_insights["frequent_hour_slot"] = customer_df.order_placed_hour.mode().values[0]
    
    customer_insights["avg_spend_per_day"] = (customer_df.groupby("order_date").total_cost.sum().sum())/date_range
    customer_insights["avg_order_per_day"] = (customer_df.groupby("order_date").order_id.count().sum())/date_range
#     customer_insights["buying_pattern"] = [i  for i in customer_df[customer_df.index.values[-5:]]]
#     customer_insights["day_of_week_graph"] = [i/number_of_weeks for i in customer_df.groupby("order_day_of_week")["total_cost"].sum()]
#     customer_insights["week_of_month_graph"] = [i/number_of_months for i in customer_df.groupby("week_of_month")["total_cost"].sum()]

    
    customer_data.append(customer_insights)
customer_insights = pd.DataFrame([a for a in customer_data])
```
